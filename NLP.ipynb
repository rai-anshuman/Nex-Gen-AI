

     NLP Phase - 1 : TEXT Processing



```python
text = "I am feeling very very happy today!!!.. " 
print("Original text: " , text)
```

```python
# Step-1 Convert text into lowercase 
# Because : Happy!=happy
#ML will consider them as different 
text=text.lower()
print("Lowercase : " , text) 
```

```python
# Step-2 Remove punctuations 
# We remove everything except letters and spaces 
# Using re - regular expression
import re
text = re.sub(r'[^\w\s]','',text)  #  re.sub(pattern, replacement, string)
print("Without punctuation:",text)
# [^\w\s] - remove everthing except letters and spaces
```

```python
# Step-3 - Tokenization (split sentences into words)
tokens = text.split()
print("Tokens :", tokens)
```

```python
# Step-4 - Remove stopwords - i, am , is, are, the, a ,an 
stopwords = ["i", "am","is","the","a","an","very","what"]

filtered_tokens =[]
for word in tokens:
    if word not in stopwords:xx
        filtered_tokens.append(word)
print("After removing stopwords :", filtered_tokens)
```

```python
def stem(word):
    if word.endswith("ing") :
        return word[:-3]
    return word

stemmed_words =[]
for word in filtered_tokens:
    stemmed_words.append(stem(word))
print("After stemming :", stemmed_words)
```



         NLP Phase -2: Words to Vector



```python
# Why Vectors?
# because ML only understand numbers
# We must convert words into numeric form 

# Bag of Words (BoW)
```

```python
# Will help us to convert text -> count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences = [ 
    "i am good",
    "i am happy",
    "happy today",
    "i am sad"
]
```

```python
# Vectorize
vectorizer = CountVectorizer() # Creating an object. By default, CountVectorizer ignores single-character words
X = vectorizer.fit_transform(sentences)
# learn vocabulary and convert sentences to numbers
```

```python
print("Vocabulary")
print(vectorizer.get_feature_names_out())
```

```python
print("Vectors :")
print(X.toarray())
```

```python
# Here:
# Each row = sentences
# Each column = word
# Each value = count 
```

```python
sentences = [
    "i am learning python and python is good",
    "Python is a great programming language to learn",
    "i also love machine learning wiht python"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
print("Vocabulary :", vectorizer.get_feature_names_out())

print("\nMatrix :")
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences  = [
    "i am learning python and python is good",
    " python is a great programming and python is fast",
    "i also love machine learning with python"
]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
#Internally it will:
# Build vocabulary  
# Calcualate TF
# Calculate IDF = log({1+N}/{1+df})+1
# Multiple tf X idf
# Generate matrix
# impact vectorization 
```

```python
print("Vocabulary :", vectorizer.get_feature_names_out())
```

```python
print("TF - IDF Matrix :")
print(X.toarray())
```
